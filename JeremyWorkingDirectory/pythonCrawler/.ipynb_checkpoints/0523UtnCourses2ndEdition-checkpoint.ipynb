```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
url = 'https://nol.ntu.edu.tw/nol/coursesearch/search_for_02_dpt.php'
# re = requests.get(url, headers=headers)



driver = webdriver.Chrome()
driver.get(url)


elem = driver.find_element_by_id('page_cnt')
elem.clear()
elem.send_keys('2000')
elem = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/font/table[1]/tbody/tr/td/table/tbody/tr[11]/td[2]/input')
elem.click()
```

```python
import time

for i in range(1,10):
    try :
        selectXpath = '/html/body/table[2]/tbody/tr/td/font/table[2]/tbody/tr/td[2]/select/option[{}]'.format(i)
        driver.find_element_by_xpath(selectXpath).click()
        time.sleep(10)
        file = open('NtuCourses1/Main{}.txt'.format(i),'w',encoding='utf8')
        file.write(driver.page_source)
        file.close()
    except:
        print('element index out of range')
    
```

```python
import pandas as pd

def HtmlCoursesInssertMongo(FilePath):
    df_courses = pd.read_html(FilePath ,encoding='utf8')[6]
    df_courses.columns= df_courses.xs(0)
    df_courses = df_courses.drop(df_courses.index[0])
    df_courses = df_courses.where(df_courses.notnull(), None)
                                                    ##if df_courses.notnull() return true, df_courses = df_courses
                                                    ##else, df_courses=None

    ##remove escape char in the column header
    columns = []
    for column in df_courses.columns:
        columns.append(column.replace("/","").replace("查看課程大綱，請點選課程名稱",""))
    df_courses.columns = columns
     
    ##Because all cells in this table are automatically transfer to string type, so I don't have to do this myself 
    def ToDIct(Row):
        return Row.to_dict()
    CoursesDictList = df_courses.apply(ToDIct, axis=1).tolist()
    return CoursesDictList



##build path(to read) list
import os
def ConvertToAbsolutePath(filename):
    return 'NtuCourses1/'+filename
PathList = pd.Series(os.listdir('NtuCourses1')).apply(ConvertToAbsolutePath).tolist()

from pymongo import MongoClient
conn = MongoClient('localhost', 27017)
db = conn.SuperUniversityCourses
collection = db['ntu']

for path in PathList:
    print(path + "has been successfully transfered!")
    collection.insert_many(HtmlCoursesInssertMongo(path))


conn.close()
```
