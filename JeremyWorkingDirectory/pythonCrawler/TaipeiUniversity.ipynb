```python
def has_class_but_no_id(tag):
    return tag.has_attr('class') and not tag.has_attr('id')
html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title" id="333"></p>

<p class="story"></p>

<p class="story">...</p>
"""
def findall(func):
    result = []
    soup = BeautifulSoup(html_doc)
    tags = soup.select('p')  
    for tag in tags:
        if(func(tag)):
            result.append(tag)
    return result
            

print(findall(has_class_but_no_id))
    
```

```python
from bs4 import BeautifulSoup
import requests

path = 'http://mops.twse.com.tw/server-java/t164sb01?step=1&CO_ID=2330&SYEAR=2016&SSEASON=4&REPORT_ID=C'
r = requests.get(path)
type(r)
r.encoding = 'big5'
html_doc = r.text
print(html_doc)
```
