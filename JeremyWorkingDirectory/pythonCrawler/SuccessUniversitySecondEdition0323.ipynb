```python
import requests
from bs4 import UnicodeDammit
from bs4 import BeautifulSoup
from bs4 import element
import re
from urllib.request import urlopen
import time
import datetime

def ExcludeFirstRow(tag):
    return tag.name == 'tr' and len(tag.contents)==48

def Weekday(time):
    if(len(time.split(']'))>1):
        return time.split(']')[0].split('[')[1]
    else:
        return 'time'

def Start(time):
    if(len(time.split(']'))>1):
        if ('~' in time):
            return time.split(']')[1].split('~')[0]
        else:
            return time.split(']')[1]
    else:
        return 'time'

def End(time):
    if(len(time.split(']'))>1):
        if ('~' in time):
            return time.split(']')[1].split('~')[1]
        else:
            return time.split(']')[1]
    else:
        return 'time'
    
def GetSemester():
    if datetime.datetime.now().month in [1,2,3,4,5,6]:
        return 2
    else:
        return 1
    


def CopyTable(path, School, CourseList) :
    html_doc1 = urlopen(path).read()
    soup2= BeautifulSoup(html_doc1, 'lxml')

    itemcount=0
    for item in soup2.find_all(ExcludeFirstRow):
        itemcount+=1
        tag = 0 
        Row = { 'UniversityID': 5 , 
                'Year' : 105, 
                'Semester': 2, 
                'CourseName':'', 
                'Credit': 0, 
                'TeacherName': '', 
                'Weekday':0, 
                'Start': 0,
                'End':0, 
                'CoreCourse': '',
                'Classroom': '',
                'SelectLimitation': '',
                'School': School,
                'Major': ''}    
        for s in item.contents:
            tag+=1
            if (tag == 33):
                if s.string == None and len(list(s.contents))>= 4:
                    #print(itemcount, Weekday(str(s.contents[1])), Weekday(str(s.contents[3])))
                    Row['Weekday'] = (Weekday(str(s.contents[1])) +','+ Weekday(str(s.contents[3])))
                    #print(itemcount, Start(str(s.contents[1])), Start(str(s.contents[3])))
                    Row['Start'] = (Start(str(s.contents[1])) +','+ Start(str(s.contents[3])))
                    #print(itemcount, End(str(s.contents[1])), End(str(s.contents[3])))
                    Row['End'] = (End(str(s.contents[1])) +','+ End(str(s.contents[3])))
                    
                elif (len(s.contents)> 1):
                    #print(itemcount, str(Weekday(str(s.contents[1]))))
                    Row['Weekday'] = (Weekday(str(s.contents[1])))
                    #print(itemcount, str(Start(str(s.contents[1]))))
                    Row['Start'] = (Start(str(s.contents[1])))
                    #print(itemcount, str(End(str(s.contents[1]))))
                    Row['End'] = (End(str(s.contents[1])))
                    
                else:
                    #print(itemcount, str(Weekday(str(s.contents[1]))))
                    Row['Weekday'] = (Weekday(str(s.string)))
                    #print(itemcount, str(Start(str(s.contents[1]))))
                    Row['Start'] = (Start(str(s.string)))
                    #print(itemcount, str(End(str(s.contents[1]))))
                    Row['End'] = (End(str(s.string)))
                    
            elif (tag == 27):
                if s.string == None and len(s.contents)>= 7:
                    #print(len(item.contents))
                    #print(itemcount, str(s.contents[0]) +','+ str(s.contents[2])+','+str(s.contents[4])+','+str(s.contents[6]))
                    Row['TeacherName'] = str(s.contents[0]) +','+ str(s.contents[2])+','+str(s.contents[4])+','+str(s.contents[6])
                elif s.string == None and len(s.contents)>= 5:
                    #print(len(item.contents))
                    #print(itemcount, str(s.contents[0]) +','+ str(s.contents[2])+','+str(s.contents[4]))
                    Row['TeacherName'] = str(s.contents[0]) +','+ str(s.contents[2])+','+str(s.contents[4])
                elif s.string == None and len(s.contents)>= 3:
                    #print(len(item.contents))
                    #print(itemcount, str(s.contents[0]) +','+ str(s.contents[2]))
                    Row['TeacherName'] =str(s.contents[0]) +','+ str(s.contents[2])
                else:
                    #print(itemcount, s.string)
                    Row['TeacherName'] =s.string
            elif (tag == 2 ):
                Row['Major'] = str(s.string).replace(' ', '')
            elif (tag == 22):
                Row['CourseName'] = s.string
            elif (tag == 24):
                Row['CoreCourse'] = s.string
            elif (tag == 26):
                Row['Credit'] = s.string
            elif (tag == 35):
                Row['Classroom'] = str(s.string).replace(' ', '')
            elif (tag == 39):
                Row['SelectLimitation'] = str(s.string).replace(' ', '')
        CourseList.append(Row)
        print(Row)
```

```python
path = 'http://course-query.acad.ncku.edu.tw/qry/'                                                         ##Courses classification Main Webform
html_doc = urlopen(path).read()
soup = BeautifulSoup(html_doc, 'lxml')


def name_a_has_target(tag):
    return tag.has_attr('target') and tag.name == 'a'
LinkList = []

def FindSchool(tag):
    return tag.name == 'div' and tag['class'] == ['theader']


#find first link
KeyTag = soup.find(href = re.compile('dept_no=A2'))     ##Search 'dept_no=A2', Get Tag including first-wanted link, and name it KeyTag
LnkDict = {'link': str('http://course-query.acad.ncku.edu.tw/qry/' + str(KeyTag['href'])), 
           'School': str(KeyTag.find_previous(FindSchool).string) }
LinkList.append(LnkDict)


#find all next links
count = 0
for item in KeyTag.find_all_next(name_a_has_target):
    LnkDict = {'link': str('http://course-query.acad.ncku.edu.tw/qry/' + str(item['href'])), 
               'School': str(item.find_previous(FindSchool).string) }
    LinkList.append(LnkDict)
    print(LnkDict)
    print('link%a success ' %(count))
    count +=1
```

```python
CourseList = []                                                                                            ##Declare Main List
count = 0

for item in LinkList:    
    CopyTable(item['link'],item['School'] , CourseList)      
    count +=1
    print('link%a success ' %(count))
```
